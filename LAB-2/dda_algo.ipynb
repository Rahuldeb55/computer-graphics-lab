```python
import matplotlib.pyplot as plt
import math

# rounding function
def round_off(n):
    if n - int(n) < 0.5:
        return int(n)
    else:
        return int(n) + 1

# DDA Line Algorithm
def dda_line(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0

    # steps calculate
    if abs(dx) > abs(dy):
        steps = abs(dx)
    else:
        steps = abs(dy)

    # increments
    x_incr = dx / steps
    y_incr = dy / steps

    x = x0
    y = y0

    x_points = []
    y_points = []

    for i in range(steps + 1):
        x_points.append(round_off(x))
        y_points.append(round_off(y))
        x += x_incr
        y += y_incr

    return x_points, y_points

# given points
x0, y0 = 200, 180
x1, y1 = 180, 160

# function call
x_pts, y_pts = dda_line(x0, y0, x1, y1)

# plotting
plt.figure()
plt.plot(x_pts, y_pts, marker='s')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("DDA Line Drawing Algorithm")
plt.grid(True)
plt.show()

# printing pixel points
print("Pixel Points:")
for x, y in zip(x_pts, y_pts):
    print(f"({x}, {y})")
```
