```python
import matplotlib.pyplot as plt

# Function to draw line when slope m > 1
def bresenham_m_greater_1(x1, y1, x2, y2):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    # Initial decision parameter
    p = 2 * dx - dy

    x, y = x1, y1       # Starting coordinates

    px, py = [], []     # Lists for storing points

    # Loop through y coordinates
    for _ in range(dy + 1):
        px.append(x)    # Store x coordinate
        py.append(y)    # Store y coordinate

        y += 1          # Move upward

        # Decision parameter update
        if p < 0:
            p += 2 * dx
        else:
            x += 1
            p += 2 * (dx - dy)

    return px, py



x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))
x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))

# Call function
px, py = bresenham_m_greater_1(x1, y1, x2, y2)

# Print generated points
print("\nGenerated Points:")
for x, y in zip(px, py):
    print(f"({x},{y})")

# Plot line using matplotlib
plt.plot(px, py, marker='o')
plt.title("Bresenham Line Drawing (m > 1)")
plt.grid(True)
plt.show()
```
