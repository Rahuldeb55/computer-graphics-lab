```python
import matplotlib.pyplot as plt

# Function to draw line when slope m <= 1
def bresenham_m_less_equal_1(x1, y1, x2, y2):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    # Initial decision parameter
    p = 2 * dy - dx

    x, y = x1, y1       # Starting point

    px, py = [], []     # Lists to store plotted points

    # Loop through x coordinates
    for _ in range(dx + 1):
        px.append(x)    # Store x coordinate
        py.append(y)    # Store y coordinate

        x += 1          # Move to next x

        # Decision parameter check
        if p < 0:
            p += 2 * dy
        else:
            y += 1
            p += 2 * (dy - dx)

    return px, py



x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))
x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))

import matplotlib.pyplot as plt  # Library for plotting line graph

# Function to draw line when slope m <= 1
def bresenham_m_less_equal_1(x1, y1, x2, y2):
    dx = abs(x2 - x1)   # Difference in x
    dy = abs(y2 - y1)   # Difference in y

    # Initial decision parameter
    p = 2 * dy - dx

    x, y = x1, y1       # Starting point

    px, py = [], []     # Lists to store plotted points

    # Loop through x coordinates
    for _ in range(dx + 1):
        px.append(x)    # Store x coordinate
        py.append(y)    # Store y coordinate

        x += 1          # Move to next x

        # Decision parameter check
        if p < 0:
            p += 2 * dy
        else:
            y += 1
            p += 2 * (dy - dx)

    return px, py


# ---- User Input Section ----
x1 = int(input("Enter x1: "))   # Starting x
y1 = int(input("Enter y1: "))   # Starting y
x2 = int(input("Enter x2: "))   # Ending x
y2 = int(input("Enter y2: "))   # Ending y

# Call function
px, py = bresenham_m_less_equal_1(x1, y1, x2, y2)

# Display generated points
print("\nGenerated Points:")
for x, y in zip(px, py):
    print(f"({x},{y})")

# Plot the line
plt.plot(px, py, marker='o')
plt.title("Bresenham Line Drawing (m ≤ 1)")
plt.grid(True)
plt.show()

px, py = bresenham_m_less_equal_1(x1, y1, x2, y2)

# Display generated points
print("\nGenerated Points:")
for x, y in zip(px, py):
    print(f"({x},{y})")

# Plot the line
plt.plot(px, py, marker='o')
plt.title("Bresenham Line Drawing (m ≤ 1)")
plt.grid(True)
plt.show()
```
