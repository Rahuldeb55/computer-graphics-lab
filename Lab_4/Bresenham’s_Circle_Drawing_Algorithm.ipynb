**Import Library**

```python
import matplotlib.pyplot as plt
```

**Bresenham Circle Function**

```python
def bresenham_circle(xc, yc, r):
    x = 0
    y = r
    d = 3 - 2 * r

    x_points = []
    y_points = []

    while x <= y:
        # 8-way symmetry
        x_points.extend([xc + x, xc - x, xc + x, xc - x,
                         xc + y, xc - y, xc + y, xc - y])

        y_points.extend([yc + y, yc + y, yc - y, yc - y,
                         yc + x, yc + x, yc - x, yc - x])

        if d < 0:
            d = d + 4 * x + 6
        else:
            d = d + 4 * (x - y) + 10
            y -= 1

        x += 1

    return x_points, y_points
```

**Take Input**

```python
xc = int(input("Enter x-center: "))
yc = int(input("Enter y-center: "))
r = int(input("Enter radius: "))
```

**plot the circle**

```python
x_points, y_points = bresenham_circle(xc, yc, r)

plt.figure()
plt.scatter(x_points, y_points)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Bresenham's Circle Drawing Algorithm")
plt.grid(True)
plt.show()
```
