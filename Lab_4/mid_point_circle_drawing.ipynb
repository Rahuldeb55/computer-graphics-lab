**Import Library**

```python
import matplotlib.pyplot as plt
```

**Mid-Point Circle Function**

```python
def midpoint_circle(xc, yc, r):
    x = 0
    y = r
    p = 1 - r

    x_points = []
    y_points = []

    while x <= y:
        # 8-way symmetry
        x_points.extend([xc + x, xc - x, xc + x, xc - x,
                         xc + y, xc - y, xc + y, xc - y])

        y_points.extend([yc + y, yc + y, yc - y, yc - y,
                         yc + x, yc + x, yc - x, yc - x])

        if p < 0:
            p = p + 2*x + 3
        else:
            p = p + 2*x - 2*y + 5
            y -= 1

        x += 1

    return x_points, y_points
```

**Take Input**

```python
xc = int(input("Enter x-center: "))
yc = int(input("Enter y-center: "))
r = int(input("Enter radius: "))
```

**plot the circle**

```python
x_points, y_points = midpoint_circle(xc, yc, r)

plt.figure()
plt.scatter(x_points, y_points)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Mid-Point Circle Drawing Algorithm")
plt.grid(True)
plt.show()
```
